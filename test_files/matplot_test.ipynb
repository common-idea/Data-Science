```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
plt.plot(x,y,'b')
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
plt.title('X-Y plot')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
plt.title('y')
plt.show()
```

```python
fig = plt.figure()
axis = fig.add_axes([0.2,0.2,0.9,0.9])
axis.plot(x,y,'b--')
axis.set_xlabel('X-lable')
axis.set_ylabel('Y-lable')
axis.set_title('X-Y PLOT')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X-label-axis1')
axes1.set_ylabel('Y-label-axis1')
axes1.set_title('Outer')

axes2.plot(y,x,'r--')
axes2.set_xlabel('X-label-axis2')
axes2.set_ylabel('Y-label-axis2')
axes2.set_title('Inner')
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,x**2)
ax[1].plot(x,x**3)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,x/2,label='x by 2')
axes.plot(x,x/3,label='x by 3')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('X-Y Plot')
axes.legend(loc='best')
```

```python
fig = plt.plot(x,y,lw=1,ls='-',marker='o',markersize=15,markerfacecolor='yellow',markeredgewidth='3',markeredgecolor='black')
```
